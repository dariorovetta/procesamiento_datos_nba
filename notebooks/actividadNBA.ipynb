```python
# Importamos librerias necesarias
import pandas as pd
import os
import matplotlib.pyplot as plt


# Acceder a la ruta donde se encuentra los archivos
ruta = os.getcwd()

# Leemos el archivo .csv
dataNBA = pd.read_csv(ruta+r"\Jugadores NBA.csv")
```

```python
# Actividad 01
# Creamos el histograma
dataNBA['PARTIDOS_JUGADOS'].plot.hist(bins=6, edgecolor="black", color="#b9a7f7ff")

# Configuración de etiquetas
plt.title('Histograma de Partidos Jugados por Profesional')
plt.xlabel('Partidos Jugados')
plt.ylabel('Frecuencia')

plt.show()

# En el rango de 70 a 80 partidos jugados, hay una frecuencia aproximada de 1000 jugadores.
```

```python
# Actividad 02
# Creamos el histograma
dataNBA['MINUTOS_JUGADOS_PROMEDIO'].plot.hist(bins=20, edgecolor="black")

# Configuración de etiquetas
plt.title('Histograma de Minutos Jugados en Promedio\n por Profesional')
plt.xlabel('Minutos Jugados Promedio')
plt.ylabel('Frecuencia')

plt.show()

# Aproximado, 440 jugadores tuvieron menos de 10 minutos jugados en promedio.
# (Se puede visualizar en las primeras 3 columnas)
```

```python
# Actividad 03
# Creamos el boxplot
dataNBA["PUNTOS_ANOTADOS_PROMEDIO"].plot.box()

# Configuración de etiquetas
plt.title('Boxplot de Puntajes Anotados en Promedio')
plt.ylabel('Puntajes Anotados Promedio')
plt.xticks([])

plt.show()

# Pregunta 1
# Los valores atípicos se encuentran por encima de los 18 puntos.

# Pregunta 2
# Obtenemos la mediana Exacta de la serie
print("La mediana de la serie es:")
print(dataNBA["PUNTOS_ANOTADOS_PROMEDIO"].median())
```

```python
# Actividad 04
# Creamos el boxplot
dataNBA["PELOTAS_RECUPERADAS_PROMEDIO"].plot.box()

# Configuración de etiquetas
plt.title('Boxplot de Pelotas Recuperadas en Promedio')
plt.ylabel('Pelotas Recuperadas Promedio')
plt.xticks([])

plt.show()

# Pregunta 1
# El valor máximo el 3,5. (Esta dentro de los valores atípicos)

# Pregunta 2
# Obtenemos los cuartiles exactos de la serie
print("Los Cuartiles de la serie son:")
print(dataNBA["PELOTAS_RECUPERADAS_PROMEDIO"].describe())
```

```python
def plot_histograms(dataframe):
    # Obtener las columnas numéricas
    numeric_columns = dataframe.select_dtypes(include=['int', 'float'])

    # Configurar el tamaño de la figura
      # Cambia el tamaño aquí según tus necesidades

    # Iterar a través de las columnas numéricas y trazar histogramas uno debajo del otro
    for i, col in enumerate(numeric_columns.columns):
        plt.figure(figsize=(6, 3))
        dataframe[col].plot.hist(bins=20, edgecolor="black", color="#b9a7f7ff")
        plt.title(f'Histograma de {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')

    plt.tight_layout()  # Ajustar el espaciado entre gráficos
    plt.show()
```

```python
# Ejemplo de uso:
plot_histograms(dataNBA)
```

```python
def plot_boxplot(dataframe):
    # Obtener las columnas numéricas
    numeric_columns = dataframe.select_dtypes(include=['int', 'float'])

    # Configurar el tamaño de la figura
      # Cambia el tamaño aquí según tus necesidades

    # Iterar a través de las columnas numéricas y trazar histogramas uno debajo del otro
    for i, col in enumerate(numeric_columns.columns):
        plt.figure(figsize=(6, 3))
        dataframe[col].plot.box()
        plt.title(f'Histograma de {col}')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.xticks([])

    plt.tight_layout()  # Ajustar el espaciado entre gráficos
    plt.show()
```

```python
# Ejemplo de uso:
plot_boxplot(dataNBA)
```
