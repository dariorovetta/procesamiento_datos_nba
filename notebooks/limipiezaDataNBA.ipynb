```python
df = df.drop(["goals_attempted", "3_point_attempted",
               "free_throws_attempted", "defensive_rebounds", "5yrs"
               ], axis = 1).set_axis(['NOMBRE', 'PARTIDOS_JUGADOS',
                  'MINUTOS_JUGADOS_PROMEDIO', 'PUNTOS_ANOTADOS_PROMEDIO', 'GOLES_PROMEDIO',
                  'TRIPLES_PROMEDIO', 'TIROS_LIBRES_ANOTADOS_PROMEDIO', 'REBOTES_OFENSIVOS_PROMEDIO',
                  'ASISTENCIAS_PROMEDIO', 'PELOTAS_RECUPERADAS_PROMEDIO', 'BLOQUEOS_PROMEDIO', "PERDIDAS_PROMEDIO"], axis=1)
df.head(2)
```

```python
data = df.dropna(subset=["NOMBRE"])
```

```python
data.isnull().sum()
```

```python
# Se remplazan los valores nulos por 0
data = data.fillna(0) 
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
# Generar el ID único utilizando un contador
id_prefix = "ID_JUG_"
id_counter = 1

def generate_unique_id():
    global id_counter
    unique_id = f"{id_prefix}{id_counter:04d}"
    id_counter += 1
    return unique_id

# Agregar una nueva columna "ID" al DataFrame con los IDs únicos generados
data['ID_JUGADOR'] = data.apply(lambda row: generate_unique_id(), axis=1)
```

```python
# Nombre de la columna que deseas mover al principio
columna_a_mover = 'ID_JUGADOR'

# Obtener el índice de la columna que deseas mover
indice_columna = data.columns.get_loc(columna_a_mover)

# Mover la columna al principio del DataFrame
data.insert(0, columna_a_mover, data.pop(columna_a_mover))

data.head()
```

```python
# Exportamos los datos finales
data.to_excel("Jugadores NBA.xlsx", encoding="utf-8", index=False)
data.to_csv("Jugadores NBA.csv", encoding="utf-8", index=False)
```
